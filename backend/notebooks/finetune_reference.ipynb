## Fine-Tuning Adit Jain's Model

We load his pre-trained weights and train them further on our specific data.

```python
from ultralytics import YOLO

# 1. Define the path to the Reference Model
# (User must ensure the .pt file is here, downloaded from the repo's Drive link)
ref_model_path = "../models/pretrained/best.pt"

# 2. Load the Reference Model
model = YOLO(ref_model_path)

# 3. Fine-Tune (Train further)
# We use a lower learning rate (lr0) because the model is already smart.
results = model.train(
    data="../data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    lr0=0.001,  # Low learning rate to preserve existing knowledge
    project="../runs/finetune",
    name="improved_adit_model",
)
```
