```python
from ultralytics import YOLO

# 1. Load Model (Nano for speed, or load 'yolov8m.pt' for accuracy)
model = YOLO('yolov8n.pt')

# 2. Start Training
# Ensure 'data.yaml' is inside 'football_cv_project' in your Drive
results = model.train(
    data='data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    project='runs/train',
    name='football_model_v1'
)
```
